```python
## List Methods

fruits.append("orange") ## Add an item to the end
print(fruits)
```

```python
fruits.insert(1,"watermelon")
print(fruits)
```

```python
# O(n) => Python first searches for the value, then shifts elements left.
#  ## Removing the first occurance of an item
fruits.remove("banana")
print(fruits)
```

```python
## Remove and return the last element
# 
# | Operation    | Complexity |
# | ------------ | ---------- |
# | pop() (last) | **O(1)**   |
# | pop(index)   | **O(n)**   |
# Removing the last element is fast.
# Removing from the middle requires shifting.
 
popped_fruits=fruits.pop()
print(popped_fruits)
print(fruits)
```

```python
# Returns the index of the first occurrence.
index=fruits.index("cherry")
print(index)
```

```python
fruits.insert(2,"banana")
print(fruits.count("banana"))
```

##### List Comprehension

List comprehension is a short and elegant way to create a new list.

Instead of writing multiple lines using a loop, you can often write the same logic in a single expression.

[expression  |   for item   |  in iterable]

      ↑             ↑              ↑

 What to store |  Variable   |  Where to iterate


Basics Syantax            [expression for item in iterable]

with conditional logic    [expression for item in iterable if condition]

Nested List Comprehension [expression for item1 in iterable1 for item2 in iterable2]




What is extend()?

Adds multiple elements from another iterable.

list.extend(iterable)

```python
numbers = [10, 20]

numbers.extend([30, 40, 50])

print(numbers)
```

```python
### Basic List Comprehension

square=[num**2 for num in range(10)]
print(square)
```

```python
### List Comprehension with Condition
lst=[]
for i in range(10):
    if i%2==0:
        lst.append(i)

print(lst)
        
```

List Comprehension with condition


General Syntax

[expression for item in iterable if condition]

```python
even_numbers=[num for num in range(10) if num%2==0]
print(even_numbers)
```

```python
## Nested List Comphrension

lst1=[1,2,3,4]
lst2=['a','b','c','d']

pair=[[i,j] for i in lst1 for j in lst2]

print(pair)
```

```python
numbers = [1,2,3,4,5]

result = ["Even" if x%2==0 else "Odd" for x in numbers]

print(result)
```

#### Conclusion
List comprehensions are a powerful and concise way to create lists in Python. They are syntactically compact and can replace more verbose looping constructs. Understanding the syntax of list comprehensions will help you write cleaner and more efficient Python code.

```python
names = ["faraz","ali","john"]

upper = [name.upper() for name in names]

print(upper)
```

```python
words = ["Python","Java"]

length = [len(word) for word in words]
print(length)
letters = [word[0] for word in words]
print(letters)
```
