##### Introduction to Tuples
Explanation:

Tuples are ordered collections of items that are immutable.
They are similar to lists, but their immutability makes them different.


```python
## creating a tuple
empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list()
print(type(lst))
tpl=tuple()
print(type(tpl))
```

```python
numbers=tuple([1,2,3,4,5,6])
numbers
```

```python
list((1,2,3,4,5,6))
type(list)
```

```python
mixed_tuple=(1,"Hello World",3.14, True)
print(mixed_tuple)
```

```python
## Accessing Tuple Elements

numbers
```

```python
print(numbers[2])
print(numbers[-1])
```

Why is the comma necessary?

Because it's the comma that creates the tuple, not the parentheses.

```python
numbers[0:4]
```

```python
numbers[::-1]
```

```python
## Tuple Operations

concatenation_tuple=numbers + mixed_tuple
print(concatenation_tuple)
```

```python
mixed_tuple * 3
```

```python
numbers *3
```

```python
## Immutable Nature Of Tuples
## Tuples are immutable, meaning their elements cannot be changed once assigned.

lst=[1,2,3,4,5]
print(lst)

lst[1]="Faraz"
print(lst)
```

```python
numbers[1]="Faraz"
```

```python
numbers
```

```python
## Tuple Methods
print(numbers.count(1))
print(numbers.index(3))
```

```python
## Packing and Unpacking tuple
## packing => Packing means putting multiple values into one tuple.
packed_tuple=1,"Hello",3.14
print(packed_tuple)
```

```python
##unpacking a tuple => Tuple unpacking means extracting the individual values stored inside a tuple and assigning them to separate variables all in a single line of code.

student = ("Faraz", 23, "Delhi")

name, age, city = student

print(name)
print(age)
print(city)
```

```python
## Unpacking with * known as Extended Unpacking
# middle becomes a list, not a tuple.
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle) # it become the list
print(last)
```

```python
## Nested Tuple
## Nested List
lst=[[1,2,3,4],[6,7,8,9],[1,"Hello",3.14,"c"]]
print(lst)
lst[0][0:3]
```

```python
lst=[[1,2,3,4],[6,7,8,9],(1,"Hello",3.14,"c")]
lst[2][0:3]
```

```python
nested_tuple = ((1, 2, 3), ("a", "b", "c"), (True, False))

## access the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating over nested tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```

#### Conclusion
Tuples are versatile and useful in many real-world scenarios where an immutable and ordered collection of items is required. They are commonly used in data structures, function arguments and return values, and as dictionary keys. Understanding how to leverage tuples effectively can improve the efficiency and readability of your Python code.

```python
# Mutable object inside a tuple
data = (1, [2, 3], 4)
data[1].append(5)

print(data)
```

```python
# | List                      | Tuple                                                 |
# | ------------------------- | ----------------------------------------------------- |
# | Mutable                   | Immutable                                             |
# | Uses `[]`                 | Uses `()`                                             |
# | More methods              | Only `count()` and `index()`                          |
# | Slightly more memory      | Slightly less memory                                  |
# | Slightly slower           | Slightly faster                                       |
# | Cannot be dictionary keys | Can be dictionary keys (if all elements are hashable) |
```

Why Are Tuples Faster?

Because Python knows the tuple cannot change.

No need to allocate extra space for future modifications or support methods like append().

That's why tuples are slightly more memory-efficient and can be faster to iterate over.
